```python
import neurokit2 as nk
import matplotlib.pyplot as plt

%matplotlib notebook


ecg = nk.ecg_simulate(duration=60, noise=0.05)

signals, info = nk.ecg_process(ecg)

print(signals.head)
print(info.keys(), info.values())
```

Plot data over seconds

```python
fig1 = nk.ecg_plot(signals, sampling_rate=1000)
```

Plot data over samples

```python
fig2 = nk.ecg_plot(signals)
```
